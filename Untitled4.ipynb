```python
import pandas as pd
import numpy as np

def generate_realistic_recovery_biomarkers(df_session):
    """Create a realistic biomarker snapshot after a hard interval workout."""

    last_hr = float(df_session["hr"].iloc[-1])
    last_power = float(df_session["power"].iloc[-1])
    last_cadence = float(df_session["cadence"].iloc[-1]) if "cadence" in df_session else 88

    # HR drift over last 10–12 minutes
    hr_tail = df_session["hr"].tail(10)
    hr_slope = np.gradient(hr_tail).mean()

    # Make recovery inputs more realistic
    biomarker_row = {
        "heart_rate": round(last_hr * np.random.uniform(0.92, 0.98), 1),   # slightly reduced after cooldown
        "power": round(last_power * np.random.uniform(0.85, 0.95), 1),     # represents morning neuromuscular power
        "cadence": round(last_cadence * np.random.uniform(0.95, 1.05), 1),
        "hr_slope_time": round(hr_slope * np.random.uniform(0.7, 1.3), 4)  # some HR drift indicates fatigue
    }

    df_bio = pd.DataFrame([biomarker_row])
    df_bio.to_csv("recovery_biomarkers_after_intervals.csv", index=False)
    return df_bio

# Run it on your generated session:
df_bio = generate_realistic_recovery_biomarkers(df_real)
df_bio
```
